```python
import pandas as pd, numpy as np
from scipy import stats   # if available; otherwise compute manually

df = pd.read_excel('enhanced_job_role_dataset_1000 (1).xlsx')
num_cols = df.select_dtypes(include=[np.number]).columns
z = np.abs((df[num_cols] - df[num_cols].mean()) / df[num_cols].std(ddof=0))
rows_z_outliers = (z > 3).any(axis=1)
print(rows_z_outliers.sum(), 'rows with z>|3| outliers')
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
cat_cols = df.select_dtypes(include="object").columns
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f"Count of {col}")
    plt.show()
```
